```python
import json
import matplotlib.pyplot as plt
```

```python
file = open("loss.json", "r")
loss_list = json.load(file)

x_vect = []
for i in range(0, len(loss_list)):
    x_vect.append(i)

plt.scatter(x_vect, loss_list, s=0.1, c='red')

plt.xlabel('Iterations')
plt.ylabel('Training Loss')
plt.title('MSASL-101 Training loss over 45 epochs')
plt.show()
```
